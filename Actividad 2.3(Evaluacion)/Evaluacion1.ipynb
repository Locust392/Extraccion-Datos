```python
#Importamos las librerias requeridas
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df= pd.read_csv('listings.csv')
df
```

Filtro O para obtener 5 hosts variables utilizadas Anne, Borzuo, Franck, Anais, Iryna

```python
# Filtro por host_name
filtro1 = df[(df["host_name"] == "Anne") |
             (df["host_name"] == "Borzou") |
             (df["host_name"] == "Franck") |
             (df["host_name"] == "Anaïs") |
             (df["host_name"] == "Iryna")]
filtro1
```

```python
filtro1.to_csv("Filtro O")
```

Comparacion

```python
# Filtro por host que se unieron antes del año 2019
filtro2 = df[df["host_since"] < "2019-01-01"]
filtro2
```

```python
filtro2.to_csv("Filtro Comparacion")
```

Por Objeto de los hosts que respoonden a mas tardar 1 dia.

```python
# Filtro por hosts que responden "a más tardar en 1 día"
filtro3 = df[df["host_response_time"] == "within a day"]
filtro3
```

```python
filtro3.to_csv("Por Objeto")
```

Por Comparacion con al menos 5 accommodates

```python
# Filtro por registros con al menos 5 accommodates
filtro4 = df[df["accommodates"] >= 5]
filtro4
```

```python
filtro4.to_csv("Filtro Comparacion 5 Accommodates")
```

Filtro O para que los tipos de habitacion sean Entire Home/apt o Private Room

```python
filtro5 = df[(df["room_type"] == "Entire home/apt") | (df["room_type"] == "Private room")]
filtro5
```

```python
filtro5.to_csv("Filtro O para Entire Home o Private Room")
```

Por Comparacion para beds que tenga a lo muco 4 camas.

```python
# Filtro por registros con a lo mucho 4 camas
filtro6 = df[df["beds"] <= 4]
filtro6
```

```python
filtro6.to_csv("Filtro Comparacion para que tenga a lo mucho 4 camas")
```

Filtro Combinado de OR con AND, para que muestre los que tengan 1 baño o sean superhost pero que todos sean instant_bookable

```python
# Filtro combinado con el conector &
filtro7 = df[(df["instant_bookable"] == "t") & ((df["bathrooms"] == 1) | (df["host_is_superhost"] == "t"))]
filtro7
```

```python
filtro7.to_csv("Filtro OR y & para tener 1 bano y superhost condicion que sean instant_bookable")
```

FILTRO Y PARA review_scores_cleanliness MENOR A 3 Y review_scores_rating MAYOR A 3

```python
df["review_scores_cleanliness"] = pd.to_numeric(df["review_scores_cleanliness"], errors='coerce')
df["review_scores_rating"] = pd.to_numeric(df["review_scores_rating"], errors='coerce')
```

```python
filtro8 = df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtro8
```

```python
filtro8.to_csv("FILTRO Y PARA review_scores_cleanliness MENOR A 3 Y review_scores_rating MAYOR A 3")
```
