```python
#Importamos las librerias requeridas
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df= pd.read_excel('Datos de Facturación.xlsx')
df
```

```python
# Cargar el Archivo Excel
file_path = 'Datos de Facturación.xlsx'
excel_data = pd.ExcelFile(file_path)
```

```python
# Cargar la hoja especifica
df = excel_data.parse('FaACTURACIÓN 2019-2023')
```

```python
# Convertirlo a numerico, convirtiendo errores a NaN
df["CVE_CLPV"] = pd.to_numeric(df["CVE_CLPV"], errors="coerce")
```

```python
#  Eliminar filas donde CVE_CLP es NaN
df = df.dropna(subset=["CVE_CLPV"])
```

FILTRO 1-CVE_CLPV (Valores: 1000 a 2000)

```python
filtro1 = df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtro1
```

2-CVE_VEND     (Todas las claves excepto “5” y “4”)


```python
filtro2 = df[~df["CVE_VEND"].isin([5, 4])]
filtro2
```

3-FECHA_ENT   (Fechas de “28/02/2022”)

```python
filtro3 = df[df["FECHA_ENT"] == "28-02-2022"]
filtro3
```

4-CAN_TOT       (Cantidades menores a 5951.7)   o  STATUS (“E”)

```python
filtro4 = df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4
```

5-Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT

```python
filtro5 = df.iloc[ : , [0,6,7,9]]
filtro5
```

6-Solo las filas de 7001-7099

```python
filtro6= df.iloc[7000:7100,:]
filtro6
```

7-Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)

```python
filtro7 = df[df["CVE_VEND"].isin([1, 2])].set_index("CVE_VEND")[["FECHAELAB"]]
filtro7
```
